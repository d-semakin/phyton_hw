```python
from sympy.plotting import plot
from sympy import Symbol, solve, diff, evalf, solveset, S, pprint, sin, cos
```

```python
#Определить корни
x = Symbol('x')
f = sin(x)**2 - cos(x)**2
ans = solve(f, x)
print(ans)
for i in ans:
    print(i.evalf())   
```

```python
#Найти интервалы, на которых функция возрастает
diff_f = diff(f, x)
pprint(solveset(diff_f > 0, x, domain=S.Reals), use_unicode=True)
```

```python
#Найти интервалы, на которых функция убывает
pprint(solveset(diff_f < 0, x, domain=S.Reals), use_unicode=True)
```

```python
#Построить график
plot(f, show=False).show()
```

```python
#Вычислить вершину
print(diff_f)
solve(diff_f, x)
```

```python
#Определить промежутки, на котором f > 0 (Найти такие x, при которых f(х) > 0) неравенствo f(x) > 0

pprint(solveset(f > 0, x, domain=S.Reals), use_unicode=True)
```

```python
#Определить промежутки, на котором f < 0 ( Найти такие x, при которых f(х) < 0) неравенствo f(x) < 0
pprint(solveset(f < 0, x, domain=S.Reals), use_unicode=True)
```

```python
# 1.Определить корни
from scipy.optimize import brentq

f = lambda x: -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30
print(brentq(f, -1, -17))
print(brentq(f, 1, 17))
```

```python
# 4. Построить график
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
def f(x):
  return -12*x**4*np.sin(np.cos(x)) - 18*x**3+5*x**2 + 10*x - 30
x = np.arange(-10, 10, 0.1)
plt.plot(x, f(x))
plt.show()
```

```python
# Сигнала в задании не было, но это пока то не много, что у меня вроде получилось
from scipy import signal 
import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return -12*x**4*np.sin(np.cos(x)) - 18*x**3+5*x**2 + 10*x - 30
x = np.arange(-10, 10, 0.1)

y = f(x)
yre = signal.resample(y,20)
xre = np.linspace(-10, 10, len(yre), endpoint=False)
plt.plot(x,y,'b', xre,yre,'or-')
plt.show()

```
